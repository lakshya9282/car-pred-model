````python
def get_brand_name(car_name):
  car_name = car_name.split(' ')[0]
  return car_name.strip(' ')
````

````python
def clean_data(value):
  value = value.split()
  value = value.strip()
  if value == '':
    value = 0
  return float (value)
````

````python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
x_train['car_name'] = label_encoder.fit_transform(x_train['car_name'])

# Create a mapping for known labels
label_mapping = dict(zip(label_encoder.classes_, label_encoder.transform(label_encoder.classes_)))

# Map test data and assign a default for unseen labels
x_test['car_name'] = x_test['car_name'].map(label_mapping).fillna(-1).astype(int)
````

````python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()

# Combine train and test sets for encoding
combined = pd.concat([x_train['car_name'], x_test['car_name']])
label_encoder.fit(combined)

# Apply encoding
x_train['car_name'] = label_encoder.transform(x_train['car_name'])
x_test['car_name'] = label_encoder.transform(x_test['car_name'])
````

````python
model.fit(x_train,y_train)
````
